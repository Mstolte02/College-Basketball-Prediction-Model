```python
CBB_RATINGS_GLOSSARY = """
CBB Team Ratings Glossary

Relative Ranking:
Each team is ranked based on how it would be expected to perform head-to-head against other similarly ranked teams. If a team is ranked 4th, it is predicted to lose against each of the top three teams, but predicted to beat teams ranked 5th, 6th, and so on.

O-Rate:
Team Offensive Efficiency Rating reflects a team’s expected offensive efficiency. This number can be interpreted as the points per 100 possessions better than average expected when playing against an average D1 team. However, each team’s O-Rate is actually based on how its offense would perform against other similarly ranked teams. A higher rating is better.

D-Rate:
Team Defensive Efficiency Rating reflects a team’s expected defensive efficiency. This number can be interpreted as the defensive points per 100 possessions better than average expected when playing against an average D1 team. However, each team’s D-Rate is actually based on how its defense would perform against other similarly ranked teams. A higher rating is better.

Relative Rating:
Team Net Relative Rating is the sum of a team’s O-Rate and D-Rate. This rating is the ultimate measure of a team’s expected overall strength, relative to other teams ranked similarly.

Change:
Improvement in BPR over the last 30 days.

Opponent Adjust:
Measures how well each team performs above or below expectation depending on opponent strength.

Pace Adjust:
Measures how well each team performs above or below expectation in faster or slower games.

Kill Shots Per Game:
Number of 10–0 scoring runs per game.

Kill Shots Allowed Per Game:
Number of 10–0 runs allowed per game.

Kill Shots Margin Per Game:
Difference between kill shots created and allowed per game.

Total Kill Shots:
Total number of 10–0 runs during the season.

Total Kill Shots Allowed:
Total number of 10–0 runs conceded.

True Tempo:
Estimated number of possessions in a game against an average-paced D1 opponent.

Off Rank:
Rank in offensive BPR.

Def Rank:
Rank in defensive BPR.

Tempo Rank:
Rank in True Tempo.

Injury Rank:
Team ranking after accounting for currently injured players.

Roster Rank:
Estimate of overall roster strength based on player BPR values.

Resume Rank:
Ranking of a team’s season resume (similar to NET).

Home Rank:
Rank of how much better a team performs at home vs. on the road.
"""
```

```python
# %% [markdown]
# # March Madness Keys — Statistical Significance Analysis
# 
# **Two levels of analysis:**
# 1. **Team-level**: Is each individual team's performance at X targets hit 
#    significantly different from the global average?
# 2. **Metric-level (pooled)**: Across ALL teams that share a given metric key,
#    is their collective performance significantly different?

# %%
import pandas as pd
import numpy as np
from scipy import stats

# === LOAD DATA ===
filepath = '/Users/markstolte/Downloads/march-madness/evan-miya/2026_March_Madness.xlsx'

df = pd.read_excel(filepath, sheet_name='Keys_Stats')
keys_des = pd.read_excel(filepath, sheet_name='Keys_Des')

df['Wins'] = (df['Team Win Pct'] * df['Num Games']).round().astype(int)
df['Losses'] = df['Num Games'] - df['Wins']

print(f"Keys_Stats: {df.shape[0]} rows ({df['Team'].nunique()} teams × {df['Targets Hit'].nunique()} target levels)")
print(f"Keys_Des: {keys_des.shape[0]} rows ({keys_des['Team'].nunique()} teams × 3 keys each)")
print(f"Unique metrics: {keys_des['Metric'].nunique()}")

# %% [markdown]
# ---
# ## Part 1: Team-Level Binomial Tests

# %%
team_results = []

for _, row in df.iterrows():
    n = int(row['Num Games'])
    k = int(row['Wins'])
    p0 = np.clip(row['Global Win Pct'], 0.001, 0.999)
    
    test = stats.binomtest(k, n, p0, alternative='two-sided')
    team_wp = row['Team Win Pct']
    
    team_results.append({
        'Team': row['Team'],
        'Targets Hit': row['Targets Hit'],
        'Num Games': n,
        'Wins': k,
        'Team Win Pct': team_wp,
        'Global Win Pct': row['Global Win Pct'],
        'Win Pct Diff': round(team_wp - row['Global Win Pct'], 3),
        'Direction': 'ABOVE' if team_wp > p0 else 'BELOW' if team_wp < p0 else 'EQUAL',
        'p_value': round(test.pvalue, 4),
        'Significant (p<0.05)': test.pvalue < 0.05,
        'Team MOV': row['Avg Margin of Victory'],
        'Global MOV': row['Global Avg MOV'],
        'MOV Diff': round(row['Avg Margin of Victory'] - row['Global Avg MOV'], 1),
    })

team_rdf = pd.DataFrame(team_results)

sig05 = team_rdf['Significant (p<0.05)'].sum()
print(f"Team-level: {sig05}/{len(team_rdf)} significant at p<0.05")
print(f"\nBy target level:")
for tgt in sorted(team_rdf['Targets Hit'].unique()):
    sub = team_rdf[team_rdf['Targets Hit'] == tgt]
    n_sig = sub['Significant (p<0.05)'].sum()
    print(f"  {tgt} targets: {n_sig}/{len(sub)} significant")

print(f"\nSignificant results (p<0.05):")
sig = team_rdf[team_rdf['Significant (p<0.05)']].sort_values('p_value')
print(sig[['Team', 'Targets Hit', 'Num Games', 'Team Win Pct', 'Global Win Pct',
           'Win Pct Diff', 'p_value', 'MOV Diff']].to_string(index=False))

# %% [markdown]
# **Takeaway**: At the team level, only 0-target rows are significant — sample
# sizes are too small for individual teams. This is why we pool by metric below.
# 
# ---
# ## Part 2: Pooled by Metric × Targets Hit
# 
# By grouping all teams that share a given metric key, we get much larger 
# sample sizes and dramatically more statistical power.

# %%
# Join metrics to stats (each stats row appears 3x, once per key)
merged = df.merge(keys_des[['Team', 'Metric', 'Target_Des']], on='Team', how='left')

metric_results = []

for metric in sorted(merged['Metric'].unique()):
    for tgt in sorted(merged['Targets Hit'].unique()):
        subset = merged[(merged['Metric'] == metric) & (merged['Targets Hit'] == tgt)]
        if len(subset) == 0:
            continue
        
        total_games = subset['Num Games'].sum()
        total_wins = subset['Wins'].sum()
        
        # Weighted global baseline
        wt_global_wp = np.average(subset['Global Win Pct'], weights=subset['Num Games'])
        pooled_wp = total_wins / total_games if total_games > 0 else 0
        
        pooled_mov = np.average(subset['Avg Margin of Victory'], weights=subset['Num Games'])
        wt_global_mov = np.average(subset['Global Avg MOV'], weights=subset['Num Games'])
        
        p0 = np.clip(wt_global_wp, 0.001, 0.999)
        test = stats.binomtest(int(total_wins), int(total_games), p0, alternative='two-sided')
        
        metric_results.append({
            'Metric': metric,
            'Targets Hit': tgt,
            'Num Teams': len(subset),
            'Total Games': total_games,
            'Total Wins': total_wins,
            'Pooled WP': round(pooled_wp, 3),
            'Global WP': round(wt_global_wp, 3),
            'WP Diff': round(pooled_wp - wt_global_wp, 3),
            'Direction': 'ABOVE' if pooled_wp > wt_global_wp else 'BELOW' if pooled_wp < wt_global_wp else 'EQUAL',
            'Pooled MOV': round(pooled_mov, 1),
            'Global MOV': round(wt_global_mov, 1),
            'MOV Diff': round(pooled_mov - wt_global_mov, 1),
            'p_value': round(test.pvalue, 4),
            'Sig (p<0.05)': test.pvalue < 0.05,
            'Sig (p<0.10)': test.pvalue < 0.10,
        })

metric_rdf = pd.DataFrame(metric_results)

total = len(metric_rdf)
sig05 = metric_rdf['Sig (p<0.05)'].sum()
sig10 = metric_rdf['Sig (p<0.10)'].sum()

print(f"Metric-level (pooled): {sig05}/{total} significant at p<0.05 ({sig05/total*100:.0f}%)")
print(f"                       {sig10}/{total} significant at p<0.10 ({sig10/total*100:.0f}%)")
print(f"\nBy target level:")
for tgt in sorted(metric_rdf['Targets Hit'].unique()):
    sub = metric_rdf[metric_rdf['Targets Hit'] == tgt]
    n05 = sub['Sig (p<0.05)'].sum()
    print(f"  {tgt} targets: {n05}/{len(sub)} significant at p<0.05")

# %%
# === SIGNIFICANT METRIC-LEVEL RESULTS ===
sig_metrics = metric_rdf[metric_rdf['Sig (p<0.05)']].sort_values(['Targets Hit', 'p_value'])

print(f"{'='*95}")
print(f"ALL SIGNIFICANT METRIC × TARGET COMBOS (p < 0.05)")
print(f"{'='*95}")
print(sig_metrics[['Metric', 'Targets Hit', 'Num Teams', 'Total Games',
                    'Pooled WP', 'Global WP', 'WP Diff', 'p_value',
                    'MOV Diff']].to_string(index=False))

# %% [markdown]
# ---
# ## Part 3: 0-to-3 Swing by Metric
# 
# Which metrics show the biggest performance swing from hitting 
# 0 targets to hitting all 3?

# %%
swing_data = []

for metric in sorted(metric_rdf['Metric'].unique()):
    r0 = metric_rdf[(metric_rdf['Metric'] == metric) & (metric_rdf['Targets Hit'] == 0)]
    r3 = metric_rdf[(metric_rdf['Metric'] == metric) & (metric_rdf['Targets Hit'] == 3)]
    
    if len(r0) > 0 and len(r3) > 0:
        swing_data.append({
            'Metric': metric,
            'WP @ 0 Targets': r0.iloc[0]['Pooled WP'],
            'WP @ 3 Targets': r3.iloc[0]['Pooled WP'],
            'WP Swing': round(r3.iloc[0]['Pooled WP'] - r0.iloc[0]['Pooled WP'], 3),
            'MOV @ 0 Targets': r0.iloc[0]['Pooled MOV'],
            'MOV @ 3 Targets': r3.iloc[0]['Pooled MOV'],
            'MOV Swing': round(r3.iloc[0]['Pooled MOV'] - r0.iloc[0]['Pooled MOV'], 1),
            'p_value @ 0': r0.iloc[0]['p_value'],
            'p_value @ 3': r3.iloc[0]['p_value'],
        })

swing_df = pd.DataFrame(swing_data).sort_values('MOV Swing', ascending=False)

print(f"{'='*95}")
print(f"METRIC IMPACT: FULL SWING FROM 0 TARGETS → 3 TARGETS")
print(f"{'='*95}")
print(swing_df.to_string(index=False))

# %% [markdown]
# ---
# ## Part 4: Save Everything

# %%
output_dir = '/Users/markstolte/Downloads/march-madness/evan-miya/'

team_rdf.sort_values(['Team', 'Targets Hit']).to_csv(
    output_dir + 'team_level_significance.csv', index=False)

metric_rdf.sort_values(['Metric', 'Targets Hit']).to_csv(
    output_dir + 'metric_level_significance.csv', index=False)

swing_df.to_csv(output_dir + 'metric_swing_analysis.csv', index=False)

print("Saved:")
print(f"  • team_level_significance.csv    ({len(team_rdf)} rows)")
print(f"  • metric_level_significance.csv  ({len(metric_rdf)} rows)")
print(f"  • metric_swing_analysis.csv      ({len(swing_df)} rows)")
```
